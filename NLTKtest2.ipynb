```python
from nltk.corpus import stopwords
```

```python
print(stopwords.words('english'))
```

```python
clean_text= [word for word in tokenized_text if word not in stopwords.words('english')]
```

```python
from nltk.stem import PorterStemmer
stemmed_text=[PorterStemmer().stem(word) for word in clean_text]
```

```python
stemmed_text
```

```python
from nltk.stem import WordNetLemmatizer
lemmatized_text=[WordNetLemmatizer().lemmatize(word) for word in clean_text]
lemmatized_text
```
